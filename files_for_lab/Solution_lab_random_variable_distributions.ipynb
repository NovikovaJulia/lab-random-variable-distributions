```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

```python
numerical = data.select_dtypes(np.number)
numerical.columns
```

```python
# 2. Check using a distribution plot if the variables fit the theoretical
# normal or exponential distribution.
```

```python
for column_name in numerical.columns:
        plt.figure()
        sns.displot(numerical[column_name], kde=True)
```

```python
for column_name in numerical.columns:
        plt.figure()
        plt.hist(numerical[column_name], bins=30)
        plt.xlabel(column_name)
        plt.show()  
```

```python
# Columns: 'Customer Lifetime Value', 'Monthly Premium Auto', 'Number of Open Complaints',
# 'Number of Policies' look like exponential distribution.
```

```python
# Column'Total Claim Amount' looks like normal distribution.
```

```python
# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point
# changes the result.
```

```python
# creating log-transformed data (Log Transformation)
# Log Transformation: Transform the response variable from y to log(y).
data_log = np.log(numerical+1e-8) # to get rid of -inf
data_log.head()
```

```python
for column_name in data_log.columns:
        plt.figure()
        sns.displot(data_log[column_name], kde=True)
```

```python
for column_name in data_log.columns:
        plt.figure()
        plt.hist(data_log[column_name], bins=30)
        plt.xlabel(column_name)
        plt.show()  
```

```python
#create log-transformed data (Square Root Transformation)
# Square Root Transformation: Transform the response variable from y to âˆšy.
data_log_sqrt_root = np.sqrt(numerical)
data_log_sqrt_root.head()
```

```python
for column_name in data_log_sqrt_root.columns:
        plt.figure()
        sns.displot(data_log_sqrt_root[column_name], kde=True)
```

```python
for column_name in data_log_sqrt_root.columns:
        plt.figure()
        plt.hist(data_log_sqrt_root[column_name], bins=30)
        plt.xlabel(column_name)
        plt.show()  
```

```python
# By performing these transformations, the dataset becomes slightly more normally distributed
# (but not much).
```
